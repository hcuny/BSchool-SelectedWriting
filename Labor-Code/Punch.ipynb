```python
def getunique(s):
    uni=[]
    for ele in s:
        if ele in uni:
            continue
        else:
            uni.append(ele)
    return uni


# Math operation done through datetime.datetime, combine virtual year & month & day
def timediff(t1,t2):
    hour=(datetime.combine(date.min,t2) - datetime.combine(date.min, t1)).total_seconds()/3600
    if hour>0:
        return hour
    else:
        return 24+hour
```

```python
def addbyhour(df,start,end,index):
    if end.hour>start.hour:
        df.loc[index,[start.hour]]=df[start.hour][index]+(3600-60*start.minute-start.second)/3600
        for x in range(start.hour+1,end.hour):
            df.loc[index,[x]]=df[x][index]+1
        df.loc[index,[end.hour]]= df[end.hour][index]+(60*end.minute+end.second)/3600

    else:
        end.hour==start.hour
        df.loc[index,[start.hour]]=df[start.hour][index]+(60*end.minute+end.second-60*start.minute-start.second)/3600
```

```python
datelist=sorted(list(set(store1['Adj Punch Date Skey'])))
hourname=[ele for ele in range(24)]
#hourname=[ele for ele in range(24)]
target_df=pd.DataFrame(columns=['date']+hourname)

ini_index=0

for d in datelist:
    target_df.loc[ini_index,['date']]=d
    ini_index+=1
    
target_df.fillna(0, inplace=True)


index=0
for d in datelist:
    subdf1=store1[store1['Adj Punch Date Skey']==d]
    emplist=getunique(subdf1['Employee Id'])
    
    for empid in emplist:
        subdf=subdf1[subdf1['Employee Id']==empid]
        subdf=subdf.sort(columns=['Adj Punch Time']) #reset index & sort by punch time
        subdf=subdf.reset_index(drop=True)
        numrow=subdf.shape[0]
        collist=subdf['Punch Type'].tolist()
        #print(subdf)
        if numrow==2:  #Only contain start & end
            try:
                #1==1
                collist==['Shift Start','Shift End']
                start=subdf['Adj Punch Time'][0]
                end=subdf['Adj Punch Time'][1]
                #first hour                
                addbyhour(target_df,start,end,index)
            except:
                print("No Good","date:",d,"empid:",empid)
                #print(subdf)
                
        elif numrow==4:
            try:

                collist==['Shift Start', 'Meal Start', 'Meal End', 'Shift End']

                start=subdf['Adj Punch Time'][0]
                rest_s=subdf['Adj Punch Time'][1]
                rest_e=subdf['Adj Punch Time'][2]
                end=subdf['Adj Punch Time'][3]
                #Part before rest                
                addbyhour(target_df,start,rest_s,index)
                #part after rest
                addbyhour(target_df,rest_e,end,index)                                            
                
            except:
                print("bad bad")
                print("No Good--","date:",d,"empid:",empid)
                
        else:
            #if subdf.shape[0]==6:
            if collist in  [['Shift Start','Meal Start','Meal End','Meal Start','Meal End','Shift End'],
                            ['Shift Start','Meal Start','Meal End','Shift End','Shift Start','Shift End'],
                            ['Shift Start','Shift End','Shift Start','Meal Start','Meal End','Shift End']]:
                
                for j in range(3):
                    start=subdf['Adj Punch Time'][2*j]
                    end=subdf['Adj Punch Time'][2*j+1]
                    addbyhour(target_df, start, end, index)             
            else:
                print(subdf['Punch Type'])
                print("--------------------")               
        #if subdf.shape[0]!=2 and subdf.shape[0]!=4 and subdf.shape[0]!=6: print(d,empid)
    index+=1    
    
    
```

```python
datelist=sorted(list(set(store1['Adj Punch Date Skey'])))
plevel_df=pd.DataFrame(columns=['date','emp_id','sys_duration'])

index=0
for d in datelist:
    subdf1=store1[store1['Adj Punch Date Skey']==d]
    emplist=getunique(subdf1['Employee Id'])

    
    for empid in emplist:
        subdf=subdf1[subdf1['Employee Id']==empid]
        subdf=subdf.sort(columns=['Adj Punch Time']) #reset index & sort by punch time
        subdf=subdf.reset_index(drop=True)
        nrows=subdf.shape[0]
        
        ngroups=nrows//2
        
        tot=0
        
        for i in range(ngroups):
            start=subdf['Adj Punch Time'][2*i]
            end= subdf['Adj Punch Time'][2*i+1]
            tot+=timediff(start,end)
        
        plevel_df.loc[index,:]=[d,empid,tot]
        index+=1
        
```
