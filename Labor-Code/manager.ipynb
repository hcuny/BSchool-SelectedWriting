```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime, date, time, timedelta
```

```python
path_1='Y://Data//ra//labor//NSDAManagerv1.xlsx'
path_2='Y://Data//ra//labor//NSDAManagerv2.xlsx'
path_3='Y://Data//ra//labor//NSDAManagerv3.xlsx'
path_4='Y://Data//ra//labor//NSDAManagerv4.xlsx'

df_1=pd.read_excel(path_1, skiprows=2)
df_2=pd.read_excel(path_2, skiprows=2)
df_3=pd.read_excel(path_3, skiprows=2)
df_4=pd.read_excel(path_4, skiprows=2)
```

```python
store1=pd.concat([df_1[df_1['Store Id']==1],df_2[df_2['Store Id']==1],df_3[df_3['Store Id']==1],df_4[df_4['Store Id']==1]],
                 ignore_index=True) #not using original index
store1.drop_duplicates(subset=['Employee Id','Date Skey','Shift End Time','Shift Start Time'],inplace=True)
store1=store1.reset_index(drop=True)
store1
```

```python
# Designed Some Preparation Functions 

def getunique(s):
    uni=[]
    for ele in s:
        if ele in uni:
            continue
        else:
            uni.append(ele)
    return uni

def roundT(t):
    if t.second!=59:
        return time(t.hour, t.minute)
    else:
        if t.minute!=59:
            return time(t.hour, t.minute+1)
        else:
            return time((t.hour+1)%24,0) 
        
def timediff(t1,t2):
    hour=(datetime.combine(date.min,t2) - datetime.combine(date.min, t1)).total_seconds()/3600
    if hour>0:
        return hour
    else:
        return 24+hour

def getrest(s, e, intev):
    
    midpoint=datetime.combine(date.min,e)-(datetime.combine(date.min,e)-datetime.combine(date.min,s))/2
    shift=timedelta(0,intev*3600)/2
    rest_start0=midpoint-shift    
    rest_end0  =midpoint+shift
    
    if e>s: # not working to midnight of the second day
        rest_start=time(rest_start0.hour, rest_start0.minute)
        rest_end= time(rest_end0.hour, rest_end0.minute)
        return rest_start, rest_end
        
    else:
        rest_start=time((rest_start0.hour+12)%24, rest_start0.minute)
        rest_end= time((rest_end0.hour+12)%24, rest_end0.minute)
        return rest_start, rest_end
    
```

```python
#--------------------------------- Transform the Dataframe Due to some formating issue---------------------------------------

#tranform time
newstart = store1['Shift Start Time'].apply(roundT)
newend   = store1['Shift End Time'].apply(roundT)

store1['Shift End Time']=newend
store1['Shift Start Time']=newstart

# float to string
newdate=[str(int(ele)) for ele in store1['Date Skey']]
store1['Date Skey']=newdate

# store format
newstore=[str(int(ele)) for ele in store1['Store Id']]
store1['Store Id']=newstore
```

```python
#---------------------------------------------- Hourly Level Analysis ------------------------------------------------------- 
#initialize null dataframe
datelist=getunique(store1['Date Skey'])
hourname=[ele for ele in range(24)]
target_df=pd.DataFrame(columns=['date']+hourname)
index=0

for d in datelist:
    target_df.loc[index,['date']]=d
    index+=1
    
target_df.fillna(0, inplace=True)    



index=0
for d in datelist:
    
    subdf=store1[store1['Date Skey']==d]
    subindex=subdf.index
        
    for i in subindex:
        start = subdf['Shift Start Time'][i]
        end   = subdf['Shift End Time'][i]
        duration=subdf['Shift Work Duration (Hrs)'][i]
        inteval=timediff(start,end)-duration                
        rest_s, rest_e=getrest(start, end, inteval)        
 
        #take care of the first hour
        target_df.loc[index,[start.hour]]=target_df[start.hour][index]+(60-start.minute)/60
        
        #working hours before break
        if rest_s.hour>start.hour:
            for x in range(start.hour+1, rest_s.hour):
                #print(x)
                target_df.loc[index,[x]]=target_df[x][index]+1
            #count the rest start hour
            target_df.loc[index,[rest_s.hour]]=target_df[rest_s.hour][index]+(rest_s.minute)/60
        else: #already the other day
            for x in range(start.hour+1, 24):
                target_df.loc[index,[x]]=target_df[x][index]+1
            for x in range(0, rest_s.hour):
                target_df.loc[index+1,[x]]=target_df[x][index+1]+1
            target_df.loc[index+1,[rest_s.hour]]=target_df[rest_s.hour][index+1]+(rest_s.minute)/60
            
        #woring hours after break
        if end>start: #must in the same day
            #count first hour after break
            target_df.loc[index,[rest_e.hour]]=target_df[rest_e.hour][index]+(60-rest_e.minute)/60
            
            for x in range(rest_e.hour+1, end.hour):
                target_df.loc[index,[x]]=target_df[x][index]+1
            #count the last hour
            target_df.loc[index,[end.hour]]=target_df[end.hour][index]+(end.minute)/60
            
        else:
            if end>rest_e: #period after break tatally in the second day
                target_df.loc[index+1,[rest_e.hour]]=target_df[rest_e.hour][index+1]+(60-rest_e.minute)/60
                for x in range(rest_e.hour+1, end.hour):
                    target_df.loc[index+1,[x]]=target_df[x][index+1]+1
                
            else: #rest end hour in the first day, end hour in the second day
                target_df.loc[index,[rest_e.hour]]=target_df[rest_e.hour][index]+(60-rest_e.minute)/60
                
                for x in range(rest_e.hour+1, 24):
                    target_df.loc[index,[x]]=target_df[x][index]+1
                    
                for x in range(0, end.hour):
                    target_df.loc[index+1,[x]]=target_df[x][index+1]+1                   
            
            #count the last hour            
            target_df.loc[index+1,[end.hour]]=target_df[end.hour][index+1]+(end.minute)/60
                                            
        
    index+=1

print("Finished, ALL Good")
        
```

```python
datelist=getunique(store1['Date Skey'])

plevel_df=pd.DataFrame(columns=['date','emp_id','sys_duration'])

ini_index=0

#For those irregular records: night shift(and have no shift the second day), put them at the end of the dataframe first.
bottomindex=100000
#Initilize all 0
for d in store1['Date Skey']:
    plevel_df.loc[ini_index,['date']]=d
    ini_index+=1
plevel_df.fillna(0, inplace=True)  

for i in store1.index:
    empid = store1['Employee Id'][i]
    date_1  = store1['Date Skey'][i]
    start = store1['Shift Start Time'][i]
    end   = store1['Shift End Time'][i]
    
    allinteval=timediff(start,end)   
    intbefore =timediff(start,time(0,0)) 
    intafter  =timediff(time(0,0),end)
    plevel_df.loc[i,['emp_id']]=empid
    if end>start:  #most case , i.e. in the same day
        plevel_df.loc[i,['sys_duration']]= plevel_df['sys_duration'][i]+allinteval
        
    else:    
        #take care of today's
        plevel_df.loc[i,['sys_duration']]= plevel_df['sys_duration'][i]+intbefore

        nextshiftday=datelist[datelist.index(date_1)+1]
        
        t=datetime.strptime(date_1,'%Y%m%d')
        nextdate=(date(t.year,t.month,t.day)+timedelta(1)).strftime('%Y%m%d')

        subdf  =store1[store1['Date Skey']==nextdate]
        if empid in subdf['Employee Id'].tolist(): #find the next position
            a=subdf[subdf['Employee Id']==empid].index.tolist()[0]
            plevel_df.loc[a,['sys_duration']]= plevel_df['sys_duration'][a]+intafter
        else:
            if end==time(0,0):
                continue
            else:
                #insert those rows at the very bottom
                plevel_df.loc[bottomindex,:]=[nextdate,empid,intafter]
                print('irregular',date_1, nextdate, nextshiftday,start,end)
                bottomindex+=1
    
# In this measure, there's a guy working on Chrismas: error 20151224 20151225 20151226 17:00:00 00:30:00    
```
